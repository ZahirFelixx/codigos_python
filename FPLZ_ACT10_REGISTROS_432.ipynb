```python
import random

MAX = 500

class Alum:
    def __init__(self):
        self.status = 0
        self.matri = 0
        self.nombre = ""
        self.appat = ""
        self.apmat = ""
        self.edad = 0
        self.sexo = 0

def decision():
    print("\n                MENU                    ")
    print("1) Registro manual")
    print("2) Registro automatico")
    print("3) Borrar")
    print("4) Ordenar")
    print("5) Buscar")
    print("6) Imprimir")
    print("0) SALIR")
    opc = int(input("\nQue deseas hacer? Ingresa el numero correspondiente: "))
    return opc

def menu():
    vect = [Alum() for _ in range(MAX)]
    i = 0
    while True:
        opc = decision()
        if opc == 1:
            if i < MAX-1:
                vect[i] = gene_manual()
                i += 1
        elif opc == 2:
            for j in range(10):
                vect[i] = gene_auto()
                i += 1
        elif opc == 3:
            eliminar(vect, i)
        elif opc == 4:
            ord(vect, i)
        elif opc == 5:
            encontrar(vect, i)
        elif opc == 6:
            imprimir(vect, i)
        elif opc == 0:
            break

def gene_manual():
    manu = Alum()
    manu.status = 1
    manu.matri = int(input("\nDame la matricula: "))

    manu.nombre = input("\nNombre(s): ")
    manu.nombre = manu.nombre.upper()

    manu.appat = input("\nApellido paterno: ")
    manu.appat = manu.appat.upper()

    manu.apmat = input("\nApellido materno: ")
    manu.apmat = manu.apmat.upper()

    manu.edad = int(input("\nIngresa tu edad: "))

    manu.sexo = int(input("Ingresa (1-. Si eres hombre) o (2-. Si eres mujer): "))

    return manu

def gene_auto():
    aut = Alum()
    aut.status = 1
    aut.matri = random.randint(300000, 399999)
    aut.sexo = random.randint(1, 2)

    if aut.sexo == 1:
        aut.nombre = nombre_mascu()
    else:
        aut.nombre = nombre_fem()

    aut.appat = apellidos()
    aut.apmat = apellidos()
    aut.edad = random.randint(10, 60)

    return aut

def apellidos():
    apell = ["FELIX","GARCIA","VALVERDE","PERAZA","YEPIZ","CHECO","PEREZ","GUZMAN","RAMIREZ","CRUZ","GOMEZ","FLORES","MORALES","VAZQUEZ","JIMENEZ","REYES","DIAZ","ORDAZ","VICENTE","SOTO"]
    return random.choice(apell)

def nombre_mascu():
    nom = ["JORGE","MIGUEL","LUIS","COLIMAN","DANIEL","MANUEL","RAYMUNDO","ROBERTO","FELIPE","MANOLO","DIEGO","RAFAEL","ANGEL","JAVIER","ANTONIO","ALFREDO","RUBEN","DIEGO","PABLO","JONATHAN"]
    return random.choice(nom)

def nombre_fem():
    nomfem = ["SINAI","BELEN","ERIKA","ALBA","VALENTINA","POPIS","VERGONIA","RAQUEL","JUANA","ORALIA","IRIS","ZURY","GUADALUPE","DAMARIS","ANGELICA","NUSLY","LIA","MARIA","PATRICIA"]
    return random.choice(nomfem)

def imprimir(vect, n):
    print("\tNo.   Matricula       Nombre(s)       Apellido Paterno   Apellido Materno     Edad       Sexo:(1-. HOMBRE 0-. MUJER)")
    if n > 0:
        for i, alum in enumerate(vect[:n]):
            print(f"\t{i+1}       {alum.matri}        {alum.nombre}         {alum.appat}        {alum.apmat}               {alum.edad}             {alum.sexo}")
    else:
        print("Sin datos")

def busqueda(vect, n, matri):
    for i, alum in enumerate(vect[:n]):
        if matri == alum.matri:
            return i
    return -1

def encontrar(vect, n):
    matri = int(input("\nIngresa la matricula que deseas: "))
    z = busqueda(vect, n, matri)

    if z != -1:
        print(f"\nResultado Exitoso. La matricula es de: {vect[z].nombre} {vect[z].appat} {vect[z].apmat}")
    else:
        print("\nError: No hay registro de la matricula que ha ingresado")

def ord(vect, n):
    for i in range(n-1):
        for j in range(i+1, n):
            if vect[j].matri < vect[i].matri:
                vect[i].matri, vect[j].matri = vect[j].matri, vect[i].matri

def eliminar(vect, n):
    matri = int(input("\nIngresa la matricula para borrar los datos: "))
    vec = [alum for alum in vect[:n] if alum.matri != matri]

    if len(vec) == n:
        print("\nError: No hay ningun dato registrado con esa matricula\n")
    else:
        print("\nEliminando registro ")
        vect[:len(vec)] = vec

        if len(vec) > 0:
            return vec[-1]
        else:
            return Alum()

if __name__ == "__main__":
    menu()
```
