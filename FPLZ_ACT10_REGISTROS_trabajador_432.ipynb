```python
import random

MAX = 500

class empleado:
    def __init__(self):
        self.numemp = 0
        self.ingreso = ""
        self.nombre = ""
        self.appat = ""
        self.apmat = ""
        self.nacimiento = ""
        self.seg_social = 0
        self.edad = 0
        self.sexo = 0
        self.direccion = ""
        self.email = ""
        self.tel = 0

def decision():
    print("\n                MENU                    ")
    print("1) Registro manual")
    print("2) Registro automatico")
    print("3) Borrar")
    print("4) Ordenar")
    print("5) Buscar")
    print("6) Imprimir")
    print("0) SALIR")
    opc = int(input("\nQue deseas hacer? Ingresa el numero correspondiente: "))
    return opc

def menu():
    vect = [empleado() for _ in range(MAX)]
    i = 0
    while True:
        opc = decision()
        if opc == 1:
            if i < MAX-1:
                vect[i] = gene_manual()
                i += 1
        elif opc == 2:
            for j in range(10):
                vect[i] = gene_auto()
                i += 1
        elif opc == 3:
            eliminar(vect, i)
        elif opc == 4:
            ord(vect, i)
        elif opc == 5:
            encontrar(vect, i)
        elif opc == 6:
            imprimir(vect, i)
        elif opc == 0:
            break

def gene_manual():
    manu = empleado()

    manu.numemp = int(input("Ingrese su numero de empleado: "))

    manu.ingreso = input("Ingrese la fecha en la cual ingreso a trabajar: ")
    manu.ingreso = manu.ingreso.upper()

    manu.nombre = input("\nNombre(s): ")
    manu.nombre = manu.nombre.upper()

    manu.appat = input("\nApellido paterno: ")
    manu.appat = manu.appat.upper()

    manu.apmat = input("\nApellido materno: ")
    manu.apmat = manu.apmat.upper()

    manu.edad = int(input("\nIngresa tu edad: "))

    manu.sexo = int(input("Ingresa (1-. Si eres hombre) o (2-. Si eres mujer): "))

    manu.nacimiento = input("\nFecha de nacimiento (DD/MM/AAAA): ")
    manu.nacimiento = manu.nacimiento.upper()

    manu.seg_social = int(input("\nNúmero de seguro social: "))

    manu.edad = int(input("\nIngresa tu edad: "))

    manu.sexo = int(input("Ingresa (1-. Si eres hombre) o (2-. Si eres mujer): "))

    manu.direccion = input("\nDireccion: ")
    manu.direccion = manu.direccion.upper()

    manu.email = input("\nCorreo electronico: ")
    manu.email = manu.email.upper()

    manu.tel = int(input("\nNúmero de telefono: "))

    return manu

def gene_auto():
    aut = empleado()
    aut.status = 1
    aut.matri = random.randint(300000, 399999)
    aut.sexo = random.randint(1, 2)

    if aut.sexo == 1:
        aut.nombre = nombre_mascu()
    else:
        aut.nombre = nombre_fem()

    aut.appat = apellidos()
    aut.apmat = apellidos()

    aut.edad = random.randint(10, 60)

    aut.nacimiento = f"{random.randint(1, 28):02}/{random.randint(1, 12):02}/{random.randint(1950, 2003):04}"
    aut.nacimiento = aut.nacimiento.upper()

    aut.seg_social = random.randint(1000000000, 9999999999)

    aut.direccion = f"Calle {random.randint(1, 100)}, Colonia {random.choice(['Centro', 'Norte', 'Sur', 'Este', 'Oeste'])}"
    aut.direccion = aut.direccion.upper()

    aut.email = f"{aut.nombre.lower()}_{aut.appat.lower()}@dominio.com"

    aut.tel = random.randint(1000000, 9999999)


    return aut

def apellidos():
    apell = ["FELIX","GARCIA","VALVERDE","PERAZA","YEPIZ","CHECO","PEREZ","GUZMAN","RAMIREZ","CRUZ","GOMEZ","FLORES","MORALES","VAZQUEZ","JIMENEZ","REYES","DIAZ","ORDAZ","VICENTE","SOTO"]
    return random.choice(apell)

def nombre_mascu():
    nom = ["JORGE","MIGUEL","LUIS","COLIMAN","DANIEL","MANUEL","RAYMUNDO","ROBERTO","FELIPE","MANOLO","DIEGO","RAFAEL","ANGEL","JAVIER","ANTONIO","ALFREDO","RUBEN","DIEGO","PABLO","JONATHAN"]
    return random.choice(nom)

def nombre_fem():
    nomfem = ["SINAI","BELEN","ERIKA","ALBA","VALENTINA","POPIS","VERGONIA","RAQUEL","JUANA","ORALIA","IRIS","ZURY","GUADALUPE","DAMARIS","ANGELICA","NUSLY","LIA","MARIA","PATRICIA"]
    return random.choice(nomfem)

def imprimir(vect, n):
    print("\tNo.   Numero de Empleado       Nombre(s)       Apellido Paterno   Apellido Materno     Edad       Sexo:(1-. HOMBRE 2-. MUJER)   Fecha de Nacimiento   Seguro Social    Direccion           Correo Electronico               Telefono")
    if n > 0:
        for i, alum in enumerate(vect[:n]):
            print(f"\t{i+1}       {alum.numemp}        {alum.nombre}         {alum.appat}        {alum.apmat}               {alum.edad}             {alum.sexo}                   {alum.nacimiento}              {alum.seg_social}          {alum.direccion}    {alum.email}              {alum.tel}")
    else:
        print("Sin datos")

def encontrar(vect, n):
    numemp = int(input("\nIngrese el numero de empleado: "))
    z = busqueda(vect, n, numemp)

    if z != -1:
        print(f"\nResultado Exitoso. El numero es de: {vect[z].nombre} {vect[z].appat} {vect[z].apmat}")
    else:
        print("\nError: No hay registro del numero de empleado que ha ingresado")

def ord(vect, n):
    for i in range(n-1):
        for j in range(i+1, n):
            if vect[j].numemp < vect[i].numemp:
                vect[i].numemp, vect[j].numemp = vect[j].numemp, vect[i].numemp

def eliminar(vect, n):
    numemp = int(input("\nIngresa el numero de empleado para borrar los datos: "))
    vec = [alum for alum in vect[:n] if alum.numemp != numemp]

    if len(vec) == n:
        print("\nError: No hay ningun dato registrado con ese numero de empleado\n")
    else:
        print("\nEliminando registro ")
        vect[:len(vec)] = vec

        if len(vec) > 0:
            return vec[-1]
        else:
            return empleado()

if __name__ == "__main__":
    menu()
```
